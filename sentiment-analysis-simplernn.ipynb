```python
docs = ['go india',
		'india india',
		'hip hip hurray',
		'jeetega bhai jeetega india jeetega',
		'bharat mata ki jai',
		'kohli kohli',
		'sachin sachin',
		'dhoni dhoni',
		'modi ji ki jai',
		'inquilab zindabad']
```

```python
from keras.preprocessing.text import Tokenizer  #we will tokenwise each word, means convert to lower case and all other
tokenizer = Tokenizer(oov_token='<nothing>')  #if some new word will come in the input then the new word will be replaced by nothing and added to unique words
```

```python
tokenizer.fit_on_texts(docs)  #tokenizer will separate the words, index generation, upper to lower case
```

```python
tokenizer.word_index  #this shows the indexes alloted to each unique word
```

```python
tokenizer.word_counts  #this shows the count of each word
```

```python
tokenizer.document_count  #this shows the number of documents or rows
```

```python
sequences = tokenizer.texts_to_sequences(docs)  #this will convert the sentences to vectors according to the indexes
sequences
```

```python
from keras.utils import pad_sequences  #as the size of each vector is not same, padding is applied inorder to make vectors of name size
sequences = pad_sequences(sequences,padding='post')  #oadding is set to post means 0s will be applied at the end of the vectors, if can also be set to pre
sequences
```

```python
model.compile('adam','accuracy')
```

```python
from keras.datasets import imdb
from keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

```python
X_train = pad_sequences(X_train,padding='post',maxlen=50)
X_test = pad_sequences(X_test,padding='post',maxlen=50)
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Embedding(10000, 2,50))
model.add(SimpleRNN(32,return_sequences=False))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(X_train, y_train,epochs=5,validation_data=(X_test,y_test))
```
